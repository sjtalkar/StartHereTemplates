```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def drawPict(loop_count):
    df = pd.DataFrame({'X':np.random.randint(1, 11, 10), 'Y': np.random.randint(1, 11, 10)})
    plt.gca().clear()
    plt.plot(df['X'], df['Y'])  
    if loop_count == 10:
        animation.event_source.stop()


```

```python
myAnim = animation.FuncAnimation(plt.gcf(), drawPict, interval=10)
plt.show()
```

## Animation with Start and Stop toggle upon click

```python
def toggleAnimation():
    animation_flag = True

    def onClick(event):
        nonlocal animation_flag
        
        if animation_flag:
            anim.event_source.stop()
            animation_flag = False
        else:
            anim.event_source.start()
            animation_flag = True
    def calc ():
        from numpy.random import rand
        n = 5
        df = pd.DataFrame (rand (n, 2), columns = ['sell', 'buy'])
        return df ['sell'], df ['buy']
    
    def plot (loop_count):
        plt.cla ()
        df_sell, df_buy = calc ()
        show = pd.concat ([df_buy, df_sell], axis = 1)
        show.plot.bar (ax = ax)
   

    fig.canvas.mpl_connect('button_press_event', onClick)

    anim = animation.FuncAnimation (fig, plot, interval = 1000)
    plt.show()
    

toggleAnimation()
```
